```python
#para treinar um modelo de machine learning, é necessário primeiro declarar uma variavel com as caracteristicas usadas para identificar o elemento la
porco1 = [0,1,0]
porco2 = [0,1,1]
porco3 = [1,1,0]
cachorro1 =[0,1,1]
cachorro2 =[1,0,1]
cachorro3 =[1,1,1]

#cria-se essa var de dados para ter um array dos elementos usados no treino da classificação
#porco 1
#cachorro 0
dados = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
classes = [1,1,1,0,0,0]

print()
```

```python
#aqui temos o import do tipo de classificação que iremos usar 
from sklearn.svm import LinearSVC

#aqui criamos uma var q instancia LinearSVC
model = LinearSVC()

#aq usamos o comando "fit" para dizer à nossa instancia "model" para aprender,
#usando os "dados" e classificando de acordo com as "classes".
model.fit(dados, classes)

print()
```

```python
#para testar nosso modelo iremos usar o exemplo do animal misterioso
animal_misterios1 = [1,1,1] #cachorro
animal_misterios2 = [1,1,0] #porco
animal_misterios3 = [0,1,1] #porco
animal_misterios4 = [1,0,0] #cachorro
animal_misterios5 = [1,0,1] #cachorro
animal_misterios6 = [0,0,1] #cachorro

teste = [animal_misterios1, animal_misterios2, animal_misterios3, animal_misterios4, animal_misterios5, animal_misterios6]

#para usar nosso modelo para identificar o que é o "animal_misterioso", usamos o comando ".predict()", passando como parametro o que queremos classificar
#o comando ".predict()" espera uma lista (array) como parametro
previsoes = model.predict(teste)
```

```python
#para testarmos os resultados do nosso modelo,m precisamos saber quais caracteristicas estamos testando de cada elemento do arra de testes
#também é com saber o que cada elemento é
#"animais" são os resultados que já sabíamos referente aos "animais misteriosos"
animais = [0,1,1,0,0,0]
#"previsoes" é uma estrutura de dados retornada pelo comando ".predict()" que é baseada em "numpy".
#por isso, ao pegar a diferença entre as previsoes e nossos resultados em "animais" usando o comando ".sum()", podemos obter o numero de resultados positivos
corretos = (animais == previsoes).sum()
```

```python
#sabendo os valores corretos da previsao, podemos fazer...
total = len(teste)
taxa_acerto = corretos/total

print("taxa acerto = ", taxa_acerto*100)
```

```python
#para nao precisar fazer calculos, ja temos no sklearn um comando que nos retorna a "accuracy" das previsoes
#temos de importá-lo
from sklearn.metrics import accuracy_score

#recebe como params primeiro os dados verdadeiros depois as previsoes
taxa_acerto2 = accuracy_score(animais, previsoes)

print ("taxa acerto = ", taxa_acerto2*100)
```
