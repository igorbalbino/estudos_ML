```python
#existe toda uma convenção sobre como devem ser os nomes das variaveis

#lendo dados da internet e classificando

#importando o modulo que usaremos para ler os dados da url
import pandas as pd

#link com a csv
url = 'https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv'

#comando do modulo pandas para ler a csv e importar para usar no codigo
data = pd.read_csv(url)

print()
```

```python
#mostra as 5 primeiras linhas da tabela
data.head()
```

```python
#para alterar o nome das colunas da tabela usamos:
#mapa para indicar qual coluna será renomeada e para qual nome
mapa = {
    "home": "principal",
    "how_it_works": "como_funciona",
    "contact": "contato",
    "bought": "comprou"
}

#comandos para renomear as "columns" baseado no que está no "mapa"
data = data.rename(columns = mapa)
```

```python
#mostrar apenas as colunas especificadas
#lê-se: dentro de uma lista do python "data" quero buscar as seguintes colunas "['home', 'how_it_works', 'contact']" (array de colunas)
x = data[['principal', 'como_funciona', 'contato']]
y = data['comprou']
```

```python
#para ver o numero de linhas e colunas da tabela importada
#esse comando é usado para, no caso de ter apenas esse conjunto de dados, dividir o conjunto afim de separar os dados de TREINO dos dados de TESTE
data.shape
```

```python
#para dividir os dados usamos:
train_x = x[:75] # train_x passa a ser tudo do elemento "x" até 75
train_y = y[:75] # train_y passa a ser tudo do elemento "y" até 75

test_x = x[75:]#test_x passa a ser tudo do elemento "x" a partir de 75
test_y = y[75:]#test_y passa a ser tudo do elemento "y" a partir de 75

train_x.shape
```

```python
#importa o modulo e o comando para realizarmos o treino e usamos os dados de train_x e train_y

from sklearn.svm import LinearSVC

model = LinearSVC()

#por algum motivo, o comando "fit" deu problema quando tentei rodar sem alterar o nome das colunas para portugues
model.fit(train_x, train_y)
```

```python
predictions = model.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score

taxa_acerto = accuracy_score(test_y, predictions) * 100

print("taxa acerto =", taxa_acerto, "%")
```

```python
#para dividir dados de treino e teste, temos um comando direto da lib sklearn
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

SEED = 20

#recebe como params "dados x", "dados y" e o tamanho que o array de teste deve receber
#para controlar a aleatoriedade do metodo "train_test_split" podemos passar, como 3º param, antes de "test_size",
#um parametro de seed, onde o comando pode se basear para fazer o split
#podemos usar, também, o param "stratify" para controlar a proporcionalidade dos conjuntos de dados "x" e "y"
train_x, test_x, train_y, test_y = train_test_split(x, y, random_state = SEED, test_size = 0.25, stratify = y)

model2 = LinearSVC()
model2.fit(train_x,train_y)

predictions = model2.predict(test_x)

taxa_acerto = accuracy_score(test_y, predictions) * 100

print(taxa_acerto)
```
